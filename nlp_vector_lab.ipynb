```python
sentence1 = "A very delicate green tea with a smooth light taste and subtle floral tones."

sentence2 = "The packaging was broken and let too much light inside the box."

sentence3 = "This product is light and easy on the stomach."
documents = [sentence1, sentence2, sentence3]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()

X = vectorizer.fit_transform(documents)

dense_matrix = X.toarray()

print(dense_matrix)
```

```python
vec_1 = dense_matrix[0]
vec_2 = dense_matrix[1]
vec_3 = dense_matrix[2]

print("Vector 1:", vec_1)
print("Vector 2:", vec_2)
print("Vector 3:", vec_3)
```

```python
import numpy as np

similarity_1_2 = np.dot(vec_1, vec_2)
similarity_1_3 = np.dot(vec_1, vec_3)

print("Similarity between Sentence 1 and Sentence 2:", similarity_1_2)
print("Similarity between Sentence 1 and Sentence 3:", similarity_1_3)
```

```python
print("Vocabulary:")
print(vectorizer.get_feature_names_out())
```

```python
import pandas as pd

feature_names = vectorizer.get_feature_names_out()

matrix_df = pd.DataFrame(
    dense_matrix,
    columns=feature_names,
    index=["Sentence 1", "Sentence 2", "Sentence 3"]
)

print(matrix_df)
```

**Conclusion:**
CountVectorizer turns words into numbers based on how often they appear, but it doesn't understand their meaning.
